```python
import scipy.misc
import numpy as np
from PIL import Image
```

```python
def charger(path):
    return scipy.misc.imread(path, flatten=True)
```

```python
mat=charger("maha.jpg")
```

```python
mat.shape, mat.size, mat
```

```python
def img_tab(img):
    #tab=np.zeros(img.size)
    tab=[]
    for i in range(img.shape[0]):
        #np.append(tab,img[i])
        tab.extend(img[i])
    return tab
```

```python
tab=img_tab(mat)
tab
```

```python
def tab_mg(tab, dim):
    tab_arr=np.array(tab)
    img=tab_arr.reshape(dim,dim)#.astype(np.int32)
    return img
```

```python
tab_img=tab_mg(tab, 600)
tab_img
```

```python
def sauvegarde(tab_dim, path):
   # scipy.misc.imsave(path, tab_dim)
    im = Image.fromarray(tab_dim)
    im.save(path)
```

```python
type(tab_img), tab_img.dtype
```

```python
sauvegarde('twit.png', tab_img)
```
