```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
mat=[]
for i in range (500) :
    a= random.normalvariate(10,4)
    b= random.normalvariate(10,4)
    c= random.normalvariate(10,4)
    mat.append([a,b,c])

    
```

```python
mat
```

```python
for i in range(100):
    
    plt.scatter(mat[i][0],mat[i][1],mat[i][2])
```

```python
 from sklearn import preprocessing
```

```python
lscale= preprocessing.scale(mat)
lscale
```

```python
lcov= np.cov(lscale.T)
```

```python
lcov
```

```python
m = np.corrcoef(lscale.T)
```

```python
m
```

```python
val,vec = np.linalg.eig(m)
```

```python
sum(val)
```

```python
vec
```
